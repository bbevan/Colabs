# Analysis via Pivots
 
Most of the data is qualitative in nature, meaning non-numerical. However, using Pivots, we can create numerical summaries of the data by counting or summing certain fields within subcategories of the raw data. 

For example, we can count the number of `Opp Deleted` "Yes" tags by Auto-Referral names, Divisions, Regions, etc. We can also take tallies of the number of applications by Region and Division.

## Number of Applicants by Division

```
# Number of Referrals by Division
first_pivot = lambda a: pd.pivot_table(data, index=a, values=["Account Name"],aggfunc=len ).sort_values(by="Account Name", ascending=False)
first_pivot("Division Name")
```

## Number of Applicants by Region

```
# Number of Referrals by Region
first_pivot("Region Name")
```

## Number of Applicants by Auto-Referral Name

```
# Number of Referrals by Auto-Referral name
first_pivot("Auto-Refer To Name")
```

## Number of Applicants by Global Position Name

```
# Number of Referrals by Global Position Name
first_pivot("Global Position Name")
```

# `Opp Deleted` Counts

One of the metrics we've spoken about is the `Opp Deleted` column. Let's try to break this down by pivot tables as well.

## Opp Deleted by Division

```
# Pivoting function
second_pivot = lambda a: pd.pivot_table(data, index=a, values=["Opp Deleted"],aggfunc=pd.DataFrame.count).sort_values(by="Opp Deleted", ascending=False)

second_pivot("Division Name")
```

## Opp Deleted by Region

```
second_pivot("Region Name")
```

## Opp Deleted by Auto-Referral Name

```
second_pivot("Auto-Refer To Name")
```

# Opp Deleted by Global Position Name

```
second_pivot("Global Position Name")
```
